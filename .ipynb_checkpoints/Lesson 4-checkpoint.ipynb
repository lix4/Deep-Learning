```python
y = np.array([128,123,129,143,132,142,112,118,108,119])
x = np.array([67,67,72,69,69,70,67,67,66,68])
```

```python
def MSE(w,b,x,y):
    return ((w*x+b-y)*(w*x+b-y)).mean()
```

```python
W = np.linspace(-1,1,101)
B= np.linspace(0,200,101)
[WW,BB]=np.meshgrid(W,B)
```

```python
mse = np.zeros((101,101))
```

```python
for i in range(101):
    for j in range(101):
        mse[i,j]=MSE(WW[i,j],BB[i,j],x,y)
plt.contour(WW,BB,mse,100)
```

```python
#standarized data
x=(x-x.mean())/(x.std())
x
```
