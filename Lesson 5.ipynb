```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f1(x1,x2):
    return x1+x1*x2+x2*x2
```

```python
X1 = np.linspace(-10,10,101)
X2= np.linspace(-10,10,101)
[X1X1,X2X2]=np.meshgrid(X1,X2)

f1_value=np.zeros((101,101))
```

```python
for i in range(101):
    for j in range(101):
        f1_value[i,j]=f1(X1X1[i,j],X2X2[i,j])
plt.contour(X1X1,X2X2,f1_value,100) #Saddle point
```

```python
def f2(x1, x2):
    return x1*x1+x1*x2+x2*x2
```

```python
a=np.linspace(0,1,101)
x1=np.array([-5,10])
x2=np.array([-5,-5])
y=np.zeros(101)
for k in range(101):
    x=a[k]*x1+(1-a[k])*x2
    y[k]=f1(x[0],x[1])
plt.plot(a,y)
```
