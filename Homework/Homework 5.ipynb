```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('.\AdjustedClosingPrices.csv')
df.head(5)
```

```python
df['Date'] = pd.to_datetime(df.Date)
```

```python
df = df.set_index('Date')
df.head(4)
```

```python
df.isnull().sum().sum()
```

```python
df1 = df.replace(0, np.nan)
```

```python
# fill the missing value
# axis = 0 is add by column 1 means filling by rows
df2 = df1.fillna(method='ffill',axis=0).fillna(method='bfill',axis=0)
df2.isnull().sum().sum()
```

```python
# If the price is 0, change that to 
df_log = np.log(df2)
df_log.isnull().sum().sum()
```

```python
df_r = df_log - df_log.shift()
```

```python
df_r.isnull().sum().sum()
```

```python
# drop the first row
df_r1 = df_r.drop(df_r.index[0])
df_r1.isnull().sum().sum()
```

```python
training_data = df_r1[0:3337]
training_data.shape

testing_data = df_r1[3338:4172]
testing_data.shape
```

```python
today_s = training_data.shift(periods=1)[2:]
tomorrow_s = training_data[2:]
```

```python
today_s.shape, tomorrow_s.shape
```

```python
today_s
```

```python
# initialize parameters
w = tf.Variable(tf.zeros([467, 467]), name="w")
b = tf.Variable(tf.zeros([1, 467]), name="b")
```

```python
x = tf.placeholder(tf.float32, shape=(3335, 467))
y = tf.placeholder(tf.float32, shape=(3335, 467))
```

```python
test_today_s = testing_data.shift(periods=1)[2:]
test_tomorrow_s = testing_data[2:]
```

```python
x_t = tf.placeholder(tf.float32, shape=(832, 467))
y_t = tf.placeholder(tf.float32, shape=(832, 467))
```

```python
loss = tf.reduce_mean(tf.square(y - (tf.matmul(x, w) + b)))
loss_test = tf.reduce_mean(tf.square(y_t - (tf.matmul(x_t, w) + b)))
```

```python
# define optimizer
optimizer = tf.train.AdamOptimizer()
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
MAXSTEPS = 1000
MOD=100
#initialize all variables
sess.run(init)
```

```python
for step in range(MAXSTEPS+1):
    (_, wopt, bopt, loss_opt,loss_t) = sess.run([train, w , b, loss, loss_test], feed_dict={x:today_s, y:tomorrow_s, x_t:test_today_s, y_t:test_tomorrow_s})
    if (step % MOD) == 0:
#         error = tf.reduce_mean(tf.square(test_tomorrow_s - (tf.matmul(test_today_s,wopt) + bopt)))
        print(step, loss_opt, loss_t)
```
