```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('./Boston_Home_Prices.csv')
df.head(3)
```

```python
features = df.iloc[:,0:13].values
prices = df.iloc[:,13].values.reshape(506,1)

print(features.shape)
print(prices.shape)
print(features)
```

```python
features_s = (features -features.mean()) / features.std()
features_s[:,3] = features[:,3]
features_s
```

```python
x = tf.placeholder(dtype='float32',shape=(506,13))
y = tf.placeholder(dtype = 'float32', shape=(506,1))
```

```python
W = tf.Variable(np.ones((13,1)),dtype='float32')
b = tf.Variable(1,dtype='float32')
```

```python
# define MSE
loss = tf.reduce_mean(tf.square(tf.matmul(x,W) + b - y))
```

```python
STEP_SIZE = 0.005
```

```python
optimizer = tf.train.GradientDescentOptimizer(STEP_SIZE).minimize(loss)

yp = tf.matmul(x,W) + b

init = tf.global_variables_initializer()
sess = tf.Session()
```

```python
sess.run(init)
MAXSTEPS = 100000
MOD = MAXSTEPS/100
for step in range(MAXSTEPS+1):
    (_,MSE)=sess.run([optimizer,loss],feed_dict={x:features_s,y:prices})
    if (step % MOD)==0:
        print('step = %-5d MSE = %-10f' % (step,MSE))
```

```python
prices.var()
```

```python
prices_pred = sess.run(yp,feed_dict={x:features_s})
```

```python
df['medv_pred'] = prices_pred
df.head(3)
```
